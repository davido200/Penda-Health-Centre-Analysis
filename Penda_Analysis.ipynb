# 7.Calculate the Net Promoter Score (NPS) in Q3 2022. (Please note that valid NPS scores range from 0 to 10).

```python
df['VisitDateTime'] = pd.to_datetime(df['VisitDateTime'], format='%d/%m/%Y %H:%M')
q3_data = df[(df['VisitDateTime'] >= '2022-07-01') & (df['VisitDateTime'] <= '2022-09-30')]
q3_data
```

```python
# Calculate NPS for Q3 2022
promoters = q3_data[q3_data['NPS Score'] >= 9].shape[0]
detractors = q3_data[q3_data['NPS Score']<= 6].shape[0]
```

```python
# Calculate NPS score
nps_score = (promoters - detractors) / q3_data.shape[0] * 100
```

```python
print(f"The Net Promote Score (NPS) is :{nps_score:.2f}")
```

# 8.In 2022 what proportion of visits in Penda Health were second visits? (A first visit is when a patient visits penda Health for the very first time, a second visit is when the same patient visit again I.e. a second time and so on)

```python
# Convert 'VisitDateTime' column to datetime format
df['VisitDateTime'] = pd.to_datetime(df['VisitDateTime'], format= '%d/%m/%Y %H:%M')
```

```python
# Sort the DataFrame by 'VisitDateTime' for each patient
df.sort_value(['PatientCode','VisitDateTime'],inplace=True)
```

```python
# Identify the second visits
df['second_visit'] = df.duplicated('PatientCode')
```

```python
df
```

```python
# Calculate the proportion of second visits
pro_second_visit = df['second_visit'].sum()/len(df)
print(f"Proportion of second visits in Penda Health:{pro_second_visit:.2f}")
```

# 9.What percent of visit in April 2022 happened within 30 days of the preceding visit by the same patient?

```python
df['TimeDifference'] = df.groupby('PatientCode')['VisitDateTime'].diff()

# Identify visits in April 2022
april_visits = df[(df['VisitDateTime'].dt.month == 4)]

# Identify visits within 30 days of the preceding visit
within_30_days = april_visits['TimeDifference'] <= pd.Timedelta(days=30)

# Calculate the percentage of visits within 30 days
percent_within_30_days = round(within_30_days.sum() / len(april_visits),2) * 100
print(f"Percent of visit in April 2022 happened within 30 days of the preceding visit by the same patient is:{percent_within_30_days}")
```
